```python
import matplotlib.pyplot as plt
from h5 import HDFArchive

# HDF5 ファイルから MaxEnt 結果と omega を読み込む
with HDFArchive("results_maxent.h5", "r") as ar:
    results_latt_orb = {}
    for key in ar.keys():
        data_dict = ar[key]
        result_data = data_dict["data"]
        result_data.omega = data_dict["omega"]  # omega も復元
        results_latt_orb[key] = result_data
```

```python
import numpy as np

aomega = []
for i in range(20):
    aomega.append(
        np.array(
            results_latt_orb[f"up_{i}"]["analyzer_results"]["LineFitAnalyzer"]["A_out"]
        )
    )
dyz = aomega[0]
dxz = aomega[1]
dxy = aomega[2]
```

```python
linestyles = ["-", "--", "-.", ":", "-", "--"]
fig, ax = plt.subplots()
ax.plot(
    results_latt_orb[f"up_0"]["omega"], dyz, label="d$_{z^2}$", linestyle=linestyles[1]
)
ax.plot(
    results_latt_orb[f"up_1"]["omega"],
    dxz,
    label="d$_{x^2-y^2}$",
    linestyle=linestyles[2],
)
ax.plot(
    results_latt_orb[f"up_2"]["omega"], dxy, label="d$_{xy}$", linestyle=linestyles[3]
)
ax.set_xlim([-8, 5])
ax.legend()
```
